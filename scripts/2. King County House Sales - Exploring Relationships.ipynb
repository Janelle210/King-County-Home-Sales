# 03. Exploring Relationships

### Correlations

```python
kc_house_data.columns
```

```python
# Create a subset excluding the unwanted columns columns

kc_sub = kc_house_data[['price', 'bedrooms', 'bathrooms', 'sqft_living15','sqft_lot15','floors', 'condition', 'grade']]
```

```python
#checking subset
kc_sub.head()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))
# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(kc_sub.corr(), annot = True, ax = ax)
```

This heat map shows some interesting relationships between variables.
0: no relationship
0.1â€“0.3: weak relationship
0.3â€“0.5: moderate relationship
0.5â€“1.0: strong relationship

There is a stronger relationship between price and bathrooms, sqft living space, grade.
There is a moderate relationship between price and bedrooms.
There is a weak relationsihp between price and sqft of the lot, number of floors, condition(surprising), year built.

The bedrooms had a strong relationship to bathrooms
moderate relationship to price, sqft living space, and grade.

Bathrooms had a strong relationship to number of floors.

The square foot living space had the highest correlation to grade, meaning more sqft the higher the grade.
It also had a strong relationship between price, bedrooms, bathrooms.
